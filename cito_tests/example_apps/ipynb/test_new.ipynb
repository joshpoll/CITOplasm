```python
from ice.utils import map_async

async def fill_activity_slot(activity_slot):
    day = activity_slot["day"]
    time = activity_slot["time"]
    activity = activity_slot["activity"]
    input_text = stringify_nested_section(district_sections[activity])
    input_text = input_text[:10000]
    response = await pick_one(input_text, f"Make sure it's open {time} on {day}. Only return the name of the item.")
    return {
            "day": day,
            "time": time,
            "activity": activity,
            "response": response['AnswerDirectly'] if 'AnswerDirectly' in response else response,
        }

filled_activity_template = await map_async(
    activity_template,
    fill_activity_slot
)

for activity in filled_activity_template:
    print(activity) 
```

```python
from dataclasses import dataclass
from citoplasm.agent.agent import Agent
from citoplasm.cito_next import create_CITO_next

@dataclass
class Synopsis:
    desc: str = "Write the synopsis of the play using the title and your thoughts."

@dataclass
class CannotAnswer:
    desc: str = "Explain why you cannot answer."

answer = await create_CITO_next(
    agent=Agent(name="a playwright", desc="Given the title of a play, it is your job to write a synopsis for that title."),
    instructions="""1. Think about the context of the title and possible themes, characters, and plot elements.
2. Write a synopsis for the play based on the title.""",
    context=None,
    output_actions=[Synopsis, CannotAnswer],
    examples=None,
    input='Title: "Tragedy at Sunset on the Beach"',
    )
```

```python
print(answer)
```

```python
import re

# split the answer on ## (Title)
split_res = re.split(r'##\s*', answer)
print(len(split_res), split_res)
```

```python
def parse_sections(input_string):
    sections = re.split(r'##\s*', input_string)
    sections_dict = {}
    
    for section in sections:
        if section.strip() == '':
            continue
        
        title, *body = section.split('\n', 1)
        body = '\n'.join(body).strip()
        sections_dict[title] = body
        
    return sections_dict
```

```python
print(parse_sections(answer))
```
